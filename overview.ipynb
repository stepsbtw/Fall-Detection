# Understanding the Data

IPQM-Fall Dataset was created from inertial variables collected from 3 main sources.
- 2 Smartwatches and a Smartphone

Each one is equipped with a series of sensors.
- Accelerometer
- Barometer
- Compass
- Geomagnetic Sensor
- Gyroscope
And the list goes on...

But the variables observed will include only:

- Linear Acceleration (Accelerometer) : Measures the rate of change in the linear speed. We have data from x, y, z axes.

- Angular Acceleration (Gyroscope) : Evaluate changes in the body's angular velocity over time. We have data from x, y, z axes.

# The Neural Network Architecture

As the idea of the neural network is to be integrated with mobile devices like smartwatches and smartphones, it is interesting to keep things simple.

Taking into account the limitations of computational resources, it was experimented using **One-Dimensional Convolutional Neural Networks** (CNN1D/CONV1D).



# Feature Extraction
In addition of the components x,y,z from linear acceleration and angular acceleration

We extract the resultant force, a variable we call magnitude.

This variable encapsulates the 3 components into a single measure of force, representing the total acceleration force applied to an object, considering the direction and intensity of each component of acceleration.

$$
R = \sqrt{a^2_x + a^2_y + a^2_z}
$$

# Data Analysis and Preparation

To better understand the patterns an analysis was done.

In fall activities a peak in the intensity of the acceleration magnitude is shown when the body hits the ground.

Activities of Daily Living do not show these peaks.

So we can take a **binary class approach** (Fall Activities and Non Fall Activities), one class represents all fall activities with and without rifle, the other class represents all daily life and military operations activities.

"falls" : 1, "non falls" : 0

## Generating Datasets in the Time Domain
Data vectors containing 450 observations (left and right watches) and 1025 (smartphone on chest) were created for each fall activity.

The data related to the body in contact with the ground was descarted, keeping only data related to the body's impact.

To maintain the same dimensionality in the input data for the neural networks, the vectors of all activities was fixed at 450 and 1025 observations.
